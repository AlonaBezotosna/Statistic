## Simple Linear Regression (SLR)

**Find linear regression coefficient for explanatory y and independent variable x**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
N = 100
x = np.array([np.random.randint(50,100) for i in range(N)])
y = np.array([i+np.random.randint(80,120)-np.random.randint(-20,20) for i in x])
plt.scatter(x, y)
```

**1. Find coefficient with method polyfit**

```python
b1, b0 = np.polyfit(x, y, 1)     #b1 - slope , b0 - intercept

plt.scatter(x, y)
plt.plot(x, b0 + b1*x, color='red')
plt.show()
print(f'b0 = {b0:.2f} and b1 = {b1:.2f}')
```

**2. Find coefficient with raw formulas  (the same rezult)**

```python
sX = sum(x)
sY = sum(y)
sXY = sum([a*b for (a,b) in zip(x,y)])
sX2 = sum([a**2 for a in x])
b1_ = (N*sXY - sX*sY)/(N*sX2 -sX*sX)
b0_ = (sY - b1_*sX)/N

plt.scatter(x, y)
plt.plot(x, b0_ + b1_*x, color='green')
plt.show()
print(f'b0_ = {b0_:.2f} and b1_ = {b1_:.2f}')
```

**3. Find coefficient with OLS from statsmodels.api (the same result)**

```python
import statsmodels.api as sm
print(sm.OLS(y, sm.add_constant(x)).fit().summary())
```

**4. Find coefficient with OLS from statsmodels.formula.api  (the same result)**

```python
from statsmodels.formula.api import ols
d = {'X':x,'Y':y}
df = pd.DataFrame(d)
model = ols('y ~ x', data = df).fit()   
print(model.summary())
#All result are the same
```

### Check condition for using regression model

```python
2. For each value of X, the probability distribution of Y has the same standard deviation σ.
3. Independence: Observations are independent of each other
```

**1.Linear relationship between dependent and independent variables** 
 
 Scatterplot above shows linear correlation between x and y

**2. Homoscedasticity: The variance of residuals is the same for any value of X**

```python
fig = plt.figure(figsize=(12,8))
fig = sm.graphics.plot_regress_exog(model, 'x', fig=fig)
```

**3. Normally distributed residuals** 

Check with plots normal distribution of residuals

```python
fig = sm.qqplot(model.resid)
plt.show()
```

```python
import seaborn as sns
sns.distplot(model.resid, hist=True)
plt.show()
```

Q-Q plot  and hystogram shoes that distribution slightly differs from normal

**Check normal distribution of residuals with Jarque-Bera test**
<br>H0: S = 0, K = 3      
<br>H1: S!=0,  K!=3
<br>where S- Skewness, K - Kurtosis

In the table "OLS Regression Results" have this statistic        
Jarque-Bera (JB):  1.48  <> 0  - the distribution is not normal

**4. Residuals independence  (no autocorrelation of residuals)**

Durbin-Watson test
<br> Already have Durbin-Watson test = 2.06 in ols.summary()
<br> Ho: no autocorrelation of residuals
<br> Test value close to 2.0 means we have no autocorrelation of residuals

```python
print(model.summary())
```

The same result can get calculation this test separately from statsmodels.stats.stattools

```python
from statsmodels.stats.stattools import durbin_watson
durbin_watson(model.resid)
```

In this example we have only one predictor x and do not have to check this test.
