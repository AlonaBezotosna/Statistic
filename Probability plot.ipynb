### Example of probability plot

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt     
```

```python
# Generate some data for this demonstration.
data = stats.norm.rvs(loc=10, scale=2.5, size=1000) 
```

```python
# Fit a normal distribution to the data:
mu, std = stats.norm.fit(data)
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2, figsize=(14, 9))
stats.probplot(data, plot=ax1)

# Plot the histogram.
ax2.hist(data, bins=25, density=True, alpha=0.6, color='g')

# Plot the PDF.
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, mu, std)
ax2.plot(x, p, 'k', linewidth=2)
```
